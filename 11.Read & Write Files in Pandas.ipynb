### How to drop a column?

```python
data_clean = pd.read_csv("winequality_clean.csv")
display (data_clean.head())
```

```python
data_clean.drop("Unnamed: 0", axis = 1)

# does not update the dataframe == the cols will still be in the orignal df
```

with drop 
- axis = 1 refers to cols
- axis = 0 refers to rows

```python
data_clean.head()
```

```python
data_clean.drop("Unnamed: 0", axis = 1, inplace = True)

# inplace = True :: the changes will be reflected in the DF
```

```python
data_clean.head()
```

```python
data_clean.drop(["density", "pH"], axis = 1, inplace = True)
data_clean.head(2)
```

```python
data_clean.drop(columns = ["volatile acidity","citric acid"], inplace = True)
data_clean.head(2)
```
