##### How are 52 cards divided?

A standard deck of playing cards consists of 52 cards. All cards are divided into 4 suits. There are two black suits — spades (♠) and clubs (♣) and two red suits — hearts (♥) and diamonds (♦). In each suit there are 13 cards including a 2, 3, 4, 5, 6, 7, 8, 9, 10, a jack, a queen, a king and an ace

```python
# What is the probability of getting a jack when we draw a card from a deck of 52 cards?

# Experiment - Drawing a card from a deck of 52 cards
# Sample Space = S = set of all the cards (52)
# Event - Even Number - E = set of all jacks (4)

cards = 52   # n(S)
jacks = 4    # n(E)
jack_probability = jacks/cards

print ("The probability of drawing an jack as a percent is {}%".
 format (round(jack_probability,5)*100))
```

```python
# What is the probability of getting a number card when we draw a card from a deck of 52 cards?

cards = 52   # n(S)
number_cards = 36    # n(E)
number_cards_probability = number_cards/cards

print ("The probability of drawing an number_cards as a percent is {}%".
 format (round(number_cards_probability,5)*100))
```

```python
# What is the chance of drawing an ace after drawing a king on the first draw?

cards = 52
aces = 4

cards_drawn = 1 #  drawing a king on the first draw
cards = cards - cards_drawn
print (cards) # remaining cards drawn

aces = 4

ace_probability = aces/cards
ace_probability

print ("The chance of drawing an ace after drawing a king on the first draw : without replacement :", 
       (round(ace_probability,5)*100))

# In sampling without replacement, each sample unit of the population has only 
# one chance to be selected in the sample. 
```

```python
# What is the chance of drawing an ace after drawing a king on the first draw and 
# the king is replaced in the deck?

# When a sampling unit is drawn from a finite population and is returned to that population, 
# after its characteristic(s) have been recorded, before the next unit is drawn, the sampling is said to be “with replacement”. 

cards = 52
aces = 4

cards_drawn = 1 #  drawing a king on the first draw
# cards = cards - cards_drawn
print (cards) # remaining cards

aces = 4

ace_probability = aces/cards
ace_probability

print ("The chance of drawing an ace after drawing a king on the first draw : with replacement : ", 
       (round(ace_probability,5)*100))
```

```python
# What is the chance of drawing an ace after drawing an ace on the first draw?

cards = 52
aces = 4

cards_drawn = 1 #  drawing a ace on the first draw
cards = cards - cards_drawn
print (cards) # remaining cards

aces_drawn = 1
aces = aces - aces_drawn
print (aces) # remaining ace cards 

ace_probability = aces/cards
ace_probability

print ("The chance of drawing an ace after drawing a king on the first draw : without replacement :", 
       (round(ace_probability,5)*100))
```

### Union, Intersection, Compliment Events and Disjoint Events

```python
A = {2,4,6} # Even Outcomes
B = {3,6} # Multiples of 3
S = {1,2,3,4,5,6}
Ac = S - A # Odd Outcomes - A compliment

print (Ac)
print (A.intersection(B))
print (A.union(B))

n_A = len(A)
n_B = len(B)
n_S = len(S)
n_Ac = len(Ac)

prob_A = n_A / n_S
prob_B = n_B / n_S
prob_Ac = n_Ac / n_S

prob_AUB = len(A.union(B))/n_S
prob_AnB = len(A.intersection(B))/n_S

print ((round(prob_A,5)*100))
print ((round(prob_B,5)*100))
print ((round(prob_Ac,5)*100))
print ((round(prob_AUB,5)*100))
print ((round(prob_AnB,5)*100))
```

```python
# What is chance of drawing an ace, a king or a queen?
#                           an ace AND (a king OR a queen)    

cards = 52

aces = 4
kings = 4
queens = 4

p_aces = 4/52
p_kings = 4/52
p_queens = 4/52

# these are disjoint events. Disjoints events cannot occur together. So all intersections are 0.

ace_king_or_queen = p_aces + p_kings + p_queens

print("The chance of drawing an ace, a king or a queen",(round(ace_king_or_queen,5)*100))
```

```python
# What is the chance of drawing a heart or drawing a club?

cards = 52

hearts = 13
clubs = 13

P_hearts = 13/52
P_clubs = 13/52

P_hearts_and_clubs = 0

heartORclub = P_hearts + P_clubs - P_hearts_and_clubs

print("There is chance of drawing a heart or drawing a club ",
      (round(heartORclub,5)*100))
```

## Independent Events

if two events X and Y are not dependent on each other, P(X|Y) = P(X).

 Independent events are those events whose occurrence is not dependent on any other event. 
 For example, if we flip a coin in the air and get the outcome as Head, 
 then again if we flip the coin but this time we get the outcome as Tail. 
 In both cases, the occurrence of both events is independent of each other.
